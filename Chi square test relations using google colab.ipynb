```python
relationship1 =pd.crosstab(index=tips['Choose your gender'],columns=tips['Age'])
relationship1
```

```python
# now let's perform the chi2 test to findout the relationship betwwen Choose your gender and Age status columns.
# here we have 2 hypothesis, the NULL hypothesis says that there is no relationship between Choose your gender and Age columns
# and the alternate hypothesis says that both the Choose your gender and Age columns are related to each other.
# to justify which hypothesis is TRUE we will have a singnificance value which is the p value, also known as alpha which 
# we are setting as 0.05
# so if p< 0.05 then we will reject the NULL hypothesis and will accept the alternate hypothesis.
```

```python
# now let's perform the chi2 test
```

```python
from scipy import stats
```

```python
relationship1.iloc[0].values
```

```python
(chi2,p_value,df,_)=stats.chi2_contingency([relationship1.iloc[0].values,relationship1.iloc[1].values])
```

```python
print(chi2)
print(p_value)
print(df)
```
